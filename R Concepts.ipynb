```r
Outer_func <- function(x){
  Inner_func <- function(y){
    a <- x+y
    return (a)
  }
  return (Inner_func)
}
output <- Outer_func(3)
output(5)
```

```r
#Recursion
recurse <- function(k){
  if(k>0){
    result <- k + recurse(k-1)
  } else{
    return (result)
  }
}
print(recurse(6))
```
